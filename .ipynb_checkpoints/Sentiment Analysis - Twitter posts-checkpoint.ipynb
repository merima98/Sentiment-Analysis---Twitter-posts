```python
#It is important to import libraries which will be used in the project
import re
import pandas as pd 
import numpy as np  
import seaborn as sns 
import nltk 
```

```python
#read_scv is also known as Read a comma-separated values 
#The first step to any data science project is to import the data
#For this project will be used read_csv function from pandas
train_dataset = pd.read_csv('https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv')
train_dataset.head()
```

```python
#Read csv - Read a comma-separated values is function from pandas. In the next line of code read_csv function will be used.
test_dataset = pd.read_csv('https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/test.csv')
test_dataset.head()
```

```python
#In the next line will be combined together train_dataset and test_dataset
#append function is used to add rows from another dataset to the end of dataset
combine_dataset = train_dataset.append(test_dataset,ignore_index=True,sort=True)
```

```python
#In the next steps will be created a function that removes unwanted text 
def remove_pattern(text,pattern):
    
    # re.findall() finds the pattern i.e @user and puts it in a list for further task
    r = re.findall(pattern,text)
    
    # re.sub() removes @user from the sentences in the dataset
    for i in r:
        text = re.sub(i,"",text)
    
    return text
```

```python
#np.vectorize () - Numpy vectorization is known as a 
#function faster than ordinary loops that work with medium or large data sets.
combine_dataset['Clean_Tweets'] = np.vectorize(remove_pattern)(combine_dataset['tweet'], "@[\w]*")

combine_dataset.head()
```

```python
#In the next line will be replaced everithing that is not in range a - z and A - Z with ' '.
combine_dataset['Clean_Tweets'] = combine_dataset['Clean_Tweets'].str.replace("[^a-zA-Z#]", " ")
#Next line will show first 10 rows of previously filtered dataset
combine_dataset.head(10)
```

```python
#Splitting a string of text into tokens with requirenment len(w)>3
combine_dataset['Clean_Tweets'] = combine_dataset['Clean_Tweets'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

combine_dataset.head(10)
```

```python
#Splitting a string of text into tokens
tokenized_tweet = combine_dataset['Clean_Tweets'].apply(lambda x: x.split())

tokenized_tweet.head()
```

```python
from nltk import PorterStemmer

ps = PorterStemmer()

#Stripping the suffixes, example: example_words = ["python","pythoner","pythoning","pythoned","pythonly"]. 
#It is stripping suffixes from all words,
#and also can be seen that every word has python in the context of word. 
tokenized_tweet = tokenized_tweet.apply(lambda x: [ps.stem(i) for i in x])

tokenized_tweet.head()
```

```python
#Stitching tokens back together is shown below
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

combine_dataset['Clean_Tweets'] = tokenized_tweet
combine_dataset.head()
```
